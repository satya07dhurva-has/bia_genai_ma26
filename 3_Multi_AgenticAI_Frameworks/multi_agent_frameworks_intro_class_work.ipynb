```python
def safe_calculate(expression : str) -> str:
    """
    This function is for calcuating simple airthemetic operatins
    Args - 
        expression : Mathematical staement in the form of string
    Returns - 
        The final result typecasted as string
        
    """
    try:
        return str(eval(expression))
    except:
        return f"Error: {str(e)}"


def search_corpus(query: str) -> str:
    """
    THis function is used to look up for documents for only the topics related to agentic/multiagetic ai, pass the topic for which iformation is required.


    Args - 
    query : Topic for which the information needs to be fetched, the topic should not be descriptive

    Returns:
        The text/information for the topic queries
    """
    corpus = {
        "langgraph" : "Langgraph is used for stateful workflows, retries, and HITL",
        "crewai" : "CrewAI focused on role-based multi-agent collaboration",
        "autogen" : "Autogen enables multi-agent conversations"
    }

    query = query.lower()
    for key, value in corpus.items():
        if key in query:
            return value
    return "No relevant results found"
```

```python
search_corpus("langgraph")
```

```python
tools = [
    Tool(
        name = "calculator",
        func = safe_calculate,
        description = "This function is for calcuating simple airthemetic operatins"
    ),
    Tool(
        name = "search",
        func = search_corpus,
        description="THis function is used to look up for documents for only the topics related to agentic/multiagetic ai, pass the topic name only, do not keep it descriptive, for which information is required"
    )

]
```

```python
model = ChatOpenAI(model = "gpt-4o-mini",
                   temperature=0)
agent = create_agent(model = model, tools = tools)
```

```python
response = agent.invoke({
    "messages": [
        {"role": "system", "content" : "Use only tools response to answer the queries, if the tool is not able to respond, then you do not have the content, respond back 'I don't know'"},
        {"role": "user", "content": "What is 8% of 20000?"}
    ]
})

response
```
